```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# set your own local path to data or download from the source
data_path = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"

df = pd.read_excel(data_path)

df.head(3)
```

```
print(f'dataset shape before cleaning is {df.shape}')
```

```
# Simple Data cleaning and formatting

df.dropna(inplace=True)
df.drop_duplicates(inplace = True)
df["InvoiceDate"] = pd.to_datetime(df.InvoiceDate, cache=True)
df["CustomerID"] = df["CustomerID"].astype(int)
df['date'] = df.InvoiceDate.astype('datetime64[D]')

print(f'dataset shape after cleaning is {df.shape}')
```

```
df['total_price'] = df['Quantity'] * df['UnitPrice']
df = df[df['total_price'] > 0]
```

```
agg_dict = {'InvoiceNo': 'nunique',
            'StockCode': 'nunique',
            'CustomerID': 'nunique',
            'total_price': 'sum',
           }

df_grouped_by_date = df.groupby(by=['date'], as_index=False).agg(agg_dict)
```
